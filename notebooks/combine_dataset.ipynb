```python
import pandas as pd
import os
import glob

def combine_parquet_to_dataframe(input_folder):
    """
    Combines all parquet files in a folder into a single pandas DataFrame.
    
    Parameters:
    input_folder (str): Path to the folder containing parquet files
    
    Returns:
    pandas.DataFrame: Combined dataframe from all parquet files
    """
    print(f"Searching for parquet files in {input_folder}...")
    
    # Get all parquet files in the folder
    parquet_files = glob.glob(os.path.join(input_folder, "*.parquet"))
    
    if not parquet_files:
        print(f"No parquet files found in {input_folder}")
        return None
    
    print(f"Found {len(parquet_files)} parquet files.")
    
    # Initialize an empty list to store individual dataframes
    dataframes = []
    
    # Process each parquet file
    for i, file_path in enumerate(parquet_files):
        try:
            print(f"Processing file {i+1}/{len(parquet_files)}: {os.path.basename(file_path)}")
            df = pd.read_parquet(file_path)
            dataframes.append(df)
            print(f"  - Shape: {df.shape}")
        except Exception as e:
            print(f"  - Error processing {file_path}: {str(e)}")
    
    if not dataframes:
        print("No data was loaded from the parquet files.")
        return None
    
    # Combine all dataframes
    print("Combining all dataframes...")
    combined_df = pd.concat(dataframes, ignore_index=True)
    print(f"Combined dataframe shape: {combined_df.shape}")
    
    return combined_df

# Usage example
input_folder = "C:\\Users\\athar\\cicddos_2019"  # Path to your folder containing parquet files

# Execute the function and store the result in a DataFrame
df_cicddos_2019 = combine_parquet_to_dataframe(input_folder)

# You can now work with the df_cicddos_2019 DataFrame
if df_cicddos_2019 is not None:
    # Display a sample of the dataframe
    print("\nSample of combined dataframe:")
    print(df_cicddos_2019.head())
    
    # Display column information
    print("\nColumns in the dataframe:")
    print(df_cicddos_2019.columns.tolist())
    
    # Display basic statistics
    print("\nDataframe info:")
    print(df_cicddos_2019.info())
```

```python
import pandas as pd
import os

def read_parquet_to_dataframe(folder_path, filename=None):
    """
    Reads a single parquet file from a folder into a pandas DataFrame.
    
    Parameters:
    folder_path (str): Path to the folder containing the parquet file
    filename (str, optional): Name of the parquet file. If None, reads the first parquet file found
    
    Returns:
    pandas.DataFrame: DataFrame created from the parquet file
    """
    try:
        # If filename is provided, use it
        if filename:
            file_path = os.path.join(folder_path, filename)
            if not os.path.exists(file_path):
                print(f"File not found: {file_path}")
                return None
        # If filename is not provided, find the first parquet file
        else:
            parquet_files = [f for f in os.listdir(folder_path) if f.endswith('.parquet')]
            if not parquet_files:
                print(f"No parquet files found in {folder_path}")
                return None
            file_path = os.path.join(folder_path, parquet_files[0])
            print(f"Using first parquet file found: {parquet_files[0]}")
        
        # Read the parquet file
        print(f"Reading parquet file: {file_path}")
        df = pd.read_parquet(file_path)
        print(f"DataFrame created with shape: {df.shape}")
        
        return df
    
    except Exception as e:
        print(f"Error reading parquet file: {str(e)}")
        return None

# Usage
folder_path = "C:\\Users\\athar\\cicdos2017_cleaned"  # Path to your folder containing the parquet file
df_cicdos_2017 = read_parquet_to_dataframe(folder_path)

# You can now work with the df_cicdos_2017 DataFrame
if df_cicdos_2017 is not None:
    # Display a sample of the dataframe
    print("\nSample of dataframe:")
    print(df_cicdos_2017.head())
    
    # Display column information
    print("\nColumns in the dataframe:")
    print(df_cicdos_2017.columns.tolist())
    
    # Display basic statistics
    print("\nDataframe info:")
    print(df_cicdos_2017.info())
```

```python
def combine_datasets_with_standard_labels(df_cicddos_2019, df_cicdos_2017, df_2018=None, df_fourth=None):
    """
    Combines multiple dataframes from different cybersecurity datasets
    and standardizes their labels to a common set.
    
    Parameters:
    df_cicddos_2019: DataFrame containing the CICIDS-DDoS 2019 dataset
    df_cicdos_2017: DataFrame containing the CICDoS 2017 dataset
    df_2018: DataFrame containing the CICIDS 2018 dataset (optional)
    df_fourth: Any additional dataframe that needs to be included (optional)
    
    Returns:
    pandas.DataFrame: Combined dataframe with standardized labels
    """
    import pandas as pd
    import numpy as np
    
    # Create copies to avoid modifying the original dataframes
    dfs = []
    
    # Process CICIDS-DDoS 2019
    if df_cicddos_2019 is not None:
        df_2019 = df_cicddos_2019.copy()
        
        # Identify the label column
        label_columns = [col for col in df_2019.columns if 'label' in col.lower() or 'class' in col.lower() or 'attack' in col.lower()]
        if label_columns:
            label_column = label_columns[0]
            print(f"Found label column in 2019 dataset: {label_column}")
            
            # Standardize labels - adjust mapping based on actual labels in your dataset
            label_map_2019 = {
                # Map your actual 2019 labels to standard labels
                # Examples (modify according to your actual data):
                'BENIGN': 'benign',
                'DDoS': 'ddos',
                'DrDoS_DNS': 'ddos',
                'DrDoS_LDAP': 'ddos',
                'DrDoS_MSSQL': 'ddos',
                'DrDoS_NetBIOS': 'ddos',
                'DrDoS_NTP': 'ddos',
                'DrDoS_SNMP': 'ddos',
                'DrDoS_SSDP': 'ddos',
                'DrDoS_UDP': 'ddos',
                'Syn': 'ddos',
                'TFTP': 'ddos',
                'UDP': 'ddos',
                'WebDDoS': 'ddos',
                'NetBIOS': 'ddos'
            }
            
            # Apply mapping - if label not in map, keep as is
            df_2019[label_column] = df_2019[label_column].apply(
                lambda x: label_map_2019.get(x, x.lower() if isinstance(x, str) else x)
            )
            
            # Add dataset source column
            df_2019['dataset_source'] = 'CICIDS-DDoS-2019'
            dfs.append(df_2019)
        else:
            print("Warning: No label column identified in 2019 dataset")
    
    # Process CICDoS 2017
    if df_cicdos_2017 is not None:
        df_2017 = df_cicdos_2017.copy()
        
        # Identify the label column
        label_columns = [col for col in df_2017.columns if 'label' in col.lower() or 'class' in col.lower() or 'attack' in col.lower()]
        if label_columns:
            label_column = label_columns[0]
            print(f"Found label column in 2017 dataset: {label_column}")
            
            # Standardize labels - adjust mapping based on actual labels in your dataset
            label_map_2017 = {
                # Map your actual 2017 labels to standard labels
                # Examples (modify according to your actual data):
                'BENIGN': 'benign',
                'DoS Hulk': 'dos',
                'DoS GoldenEye': 'dos',
                'DoS slowloris': 'dos',
                'DoS Slowhttptest': 'dos',
                'Heartbleed': 'dos'
            }
            
            # Apply mapping - if label not in map, keep as is
            df_2017[label_column] = df_2017[label_column].apply(
                lambda x: label_map_2017.get(x, x.lower() if isinstance(x, str) else x)
            )
            
            # Add dataset source column
            df_2017['dataset_source'] = 'CICDoS-2017'
            dfs.append(df_2017)
        else:
            print("Warning: No label column identified in 2017 dataset")
    
    # Process CICIDS 2018
    if df_2018 is not None:
        df2018 = df_2018.copy()
        
        # Identify the label column
        label_columns = [col for col in df2018.columns if 'label' in col.lower() or 'class' in col.lower() or 'attack' in col.lower()]
        if label_columns:
            label_column = label_columns[0]
            print(f"Found label column in 2018 dataset: {label_column}")
            
            # Standardize labels
            label_map_2018 = {
                # Map your actual 2018 labels to standard labels:
                'Benign': 'benign',
                'DoS': 'dos',
                'DDoS': 'ddos',
                'Brute Force': 'bruteforce',
                'Web Attack': 'webattack',
                'Infiltration': 'infilteration',
                'Botnet': 'botnet',
                'Port Scan': 'portscan',
                # Also handle numeric labels if they exist
                1: 'benign',
                2: 'dos',
                3: 'ddos',
                4: 'bruteforce',
                5: 'webattack',
                6: 'infilteration',
                7: 'botnet',
                8: 'portscan'
            }
            
            # Apply mapping
            df2018[label_column] = df2018[label_column].apply(
                lambda x: label_map_2018.get(x, x.lower() if isinstance(x, str) else x)
            )
            
            # Add dataset source column
            df2018['dataset_source'] = 'CICIDS-2018'
            dfs.append(df2018)
        else:
            print("Warning: No label column identified in 2018 dataset")
    
    # Process fourth dataset if provided
    if df_fourth is not None:
        df4 = df_fourth.copy()
        
        # Identify the label column
        label_columns = [col for col in df4.columns if 'label' in col.lower() or 'class' in col.lower() or 'attack' in col.lower()]
        if label_columns:
            label_column = label_columns[0]
            print(f"Found label column in fourth dataset: {label_column}")
            
            # Standardize labels - adjust mapping based on actual labels in your dataset
            label_map_4 = {
                # Map your actual labels to standard labels
                # Modify according to your actual data
            }
            
            # Apply mapping - if label not in map, keep as is
            df4[label_column] = df4[label_column].apply(
                lambda x: label_map_4.get(x, x.lower() if isinstance(x, str) else x)
            )
            
            # Add dataset source column
            df4['dataset_source'] = 'Fourth-Dataset'
            dfs.append(df4)
        else:
            print("Warning: No label column identified in fourth dataset")
    
    # Check if we have dataframes to combine
    if not dfs:
        print("Error: No valid dataframes to combine")
        return None
    
    # Identify common columns across all dataframes
    common_columns = set.intersection(*[set(df.columns) for df in dfs])
    print(f"Found {len(common_columns)} common columns across datasets")
    
    # If there are very few common columns, we might need to keep all columns
    if len(common_columns) < 5:
        print("Warning: Very few common columns. Combining all columns and filling missing values with NaN")
        combined_df = pd.concat(dfs, ignore_index=True, sort=False)
    else:
        print(f"Using common columns for combining datasets: {common_columns}")
        combined_df = pd.concat([df[list(common_columns)] for df in dfs], ignore_index=True)
    
    # Standardize the label column name across the combined dataset
    label_cols = [col for col in combined_df.columns if 'label' in col.lower() or 'class' in col.lower() or 'attack' in col.lower()]
    if label_cols:
        # Rename the first identified label column to 'label'
        combined_df = combined_df.rename(columns={label_cols[0]: 'label'})
        
        # Make sure all labels are in the required set
        valid_labels = ['benign', 'ddos', 'dos', 'botnet', 'bruteforce', 'infilteration', 'webattack', 'portscan']
        
        # Map any other label to the closest match or 'benign' if can't determine
        combined_df['label'] = combined_df['label'].apply(
            lambda x: x if x in valid_labels else 'benign'
        )
        
        # Print label distribution
        print("\nLabel distribution in combined dataset:")
        print(combined_df['label'].value_counts())
    else:
        print("Warning: No label column found in combined dataset")
    
    return combined_df

# Usage example (adjust variable names as needed)
# Assuming you have these four dataframes already loaded:
# df_cicddos_2019, df_cicdos_2017, df_2018, df_fourth

# Combine the datasets
combined_df = combine_datasets_with_standard_labels(
    df_cicddos_2019=df_cicddos_2019, 
    df_cicdos_2017=df_cicdos_2017,
    df_2018=df_2018,  # Replace with your actual third dataframe if different
    df_fourth=None    # Replace with your actual fourth dataframe if available
)

# Check the combined dataframe
if combined_df is not None:
    print(f"\nCombined dataset shape: {combined_df.shape}")
    print("\nSample of combined dataset:")
    print(combined_df.head())
    
    # Verify all labels are standardized
    print("\nVerifying labels in combined dataset:")
    print(combined_df['label'].unique())
```
